```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_parquet(r"..\data\raw\fred-observations\DGS2\2005-01-01_2026-05-04.parquet")
display(df)
df2 = pd.read_parquet(r"..\data\raw\fred-observations\DGS30\2005-01-01_2026-05-04.parquet")
display(df2)

plot_df = df.dropna(subset=["value"])
plt.figure(figsize=(10,4))
plt.plot(plot_df["date"], plot_df["value"])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plot_df2 = df2.dropna(subset=["value"])
plt.figure(figsize=(10, 4))
plt.plot(plot_df2["date"], plot_df2["value"])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
