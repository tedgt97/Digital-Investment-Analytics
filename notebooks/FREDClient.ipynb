```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_parquet(r"..\data\raw\fred-observations\FEDFUNDS\2005-01-01_2026-04-01.parquet")
display(df)

plot_df = df.dropna(subset=["value"])

plt.figure(figsize=(10,4))
plt.plot(plot_df["date"], plot_df["value"])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
