#### **Single Call**

```python
symbol = 'AAPL'
start_date = '2020-01-01'
end_date = '2025-12-13'

df_chart = client.get_chart(symbol, start_date, end_date)
df_chart
```

```python
import matplotlib.pyplot as plt

df_chart_plot = df_chart.copy()
df_chart_plot['date'] = pd.to_datetime(df_chart_plot['date'])
df_chart_plot = df_chart_plot.set_index('date')

plt.figure(figsize=(10, 4))
df_chart_plot['close'].plot()
plt.title(f'{symbol} Close Price')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
