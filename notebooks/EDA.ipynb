```python
import os
from pathlib import Path
```

```python
start = Path.cwd().resolve()
print(start)
print(start.parents[0])
print(start.parents[0] / 'config' / 'api_keys.txt')

config_file = start.parents[0] / 'config' / 'api_keys.txt'
print(config_file)
```

```python
with open(config_file, 'r') as f:
    lines = f.readlines()

    for line in lines:
        line = line.strip()
        if line and not line.startswith('#'):
            if line.startswith('FRED_API_KEY'):
                api_key = line.split('=')[1].strip()

print(lines)
print(api_key)
```
